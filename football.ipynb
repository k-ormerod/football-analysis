```python
import matplotlib.pyplot as plt
import numpy as np
from mplsoccer import Pitch, Sbopen, VerticalPitch

# import plotting style 
plt.style.use('/Users/katherineormerod/Documents/mnras copy.mplstyle.mplstyle')
```

```python
parser = Sbopen()

df, related, freeze, tactics = parser.event(3844384)

team1, team2 = df.team_name.unique()

shots = df.loc[df["type_name"] == "Shot"].set_index("id")
```

```python
pitch = Pitch(line_color = "black")
fig, ax = pitch.draw(figsize=(10, 7))
#Size of the pitch in yards (!!!)
pitchLengthX = 120
pitchWidthY = 80
#Plot the shots by looping through them.
for i,shot in shots.iterrows():
    #get the information
    x=shot['x']
    y=shot['y']
    goal=shot['outcome_name']=='Goal'
    team_name=shot['team_name']
    #set circlesize
    circleSize=2
    #plot England
    if (team_name==team1):
        if goal:
            shotCircle=plt.Circle((x,y),circleSize,color="red")
            plt.text(x+1,y-2,shot['player_name'])
        else:
            shotCircle=plt.Circle((x,y),circleSize,color="red")
            shotCircle.set_alpha(.2)
    #plot Sweden
    else:
        if goal:
            shotCircle=plt.Circle((pitchLengthX-x,pitchWidthY - y),circleSize,color="blue")
            plt.text(pitchLengthX-x+1,pitchWidthY - y - 2 ,shot['player_name'])
        else:
            shotCircle=plt.Circle((pitchLengthX-x,pitchWidthY - y),circleSize,color="blue")
            shotCircle.set_alpha(.2)
    ax.add_patch(shotCircle)
#set title
fig.suptitle("England (red) and Spain (blue) shots", fontsize = 24)
fig.set_size_inches(10, 7)
plt.show()
```

```python
#create pitch
pitch = Pitch(line_color='black')
fig, ax = pitch.grid(grid_height=0.9, title_height=0.06, axis=False,
                     endnote_height=0.04, title_space=0, endnote_space=0)
#query
mask_england = (df.type_name == 'Shot') & (df.team_name == team1)
#finding rows in the df and keeping only necessary columns
df_england = df.loc[mask_england, ['x', 'y', 'end_x', 'end_y','outcome_name', "player_name"]]

#plot them - if shot ended with Goal - alpha 1 and add name
#for England
for i, row in df_england.iterrows():
    if row["outcome_name"] == 'Goal':
    #make circle
       pitch.scatter(row.x, row.y, alpha = 1, s = 500, color = "red", ax=ax['pitch'])
       pitch.annotate(row["player_name"], (row.x + 1, row.y - 2), ax=ax['pitch'], fontsize = 12)
    else:
       pitch.scatter(row.x, row.y, alpha = 0.2, s = 500, color = "red", ax=ax['pitch'])

mask_spain = (df.type_name == 'Shot') & (df.team_name == team2)
df_spain = df.loc[mask_spain, ['x', 'y', 'outcome_name', "player_name"]]

#for Sweden we need to revert coordinates
for i, row in df_spain.iterrows():
    if row["outcome_name"] == 'Goal':
       pitch.scatter(120 - row.x, 80 - row.y, alpha = 1, s = 500, color = "blue", ax=ax['pitch'])
       pitch.annotate(row["player_name"], (120 - row.x + 1, 80 - row.y - 2), ax=ax['pitch'], fontsize = 12)
    else:
       pitch.scatter(120 - row.x, 80 - row.y, alpha = 0.2, s = 500, color = "blue", ax=ax['pitch'])

fig.suptitle("England (red) and Spain (blue) shots", fontsize = 30)
plt.show()
```

```python
import matplotlib.patheffects as path_effects
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from matplotlib.colors import LinearSegmentedColormap
from scipy.ndimage import gaussian_filter

from mplsoccer import Pitch, VerticalPitch, FontManager, Sbopen


# print the competitions available

df = sb.competitions()

comp = "FA Women's Super League"


comp_id = 37
season_id = 90

matches = sb.matches(competition_id = comp_id, season_id = season_id)

team = "Manchester City WFC"

# select home team or away team == team

# matches.loc[(matches["home_team"] == team) | (matches["away_team"] == team)]

# match_id = 3775658


#list all match ids with manchester city womens 

df = matches.loc[(matches["home_team"] == team) | (matches["away_team"] == team)]

matchids = []

for i, row in df.iterrows():
    matchids.append(row["match_id"])

parser = Sbopen()

df = pd.concat([parser.event(match_id)[0] for match_id in matchids])

# df, related, freeze, tactics = parser.event(match_id)


# team1, team2 = df.team_name.unique()

# poss = df.loc[df["type_name"] == "Possession"].set_index("id")



mask_mcwfc = (df.type_name == 'Pressure') & (df.team_name == team)

df_pressure = df.loc[mask_mcwfc, ['x', 'y']]


pitch = VerticalPitch(pitch_type='statsbomb', line_zorder=2,
              pitch_color='#22312b', line_color='#efefef', corner_arcs=True,
              goal_type = 'box')
# draw
fig, ax = pitch.draw(figsize=(6.6, 4.125))
fig.set_facecolor('#22312b')

bin_statistic = pitch.bin_statistic(df_pressure.x, df_pressure.y, statistic='count', bins=(25, 25))
bin_statistic['statistic'] = gaussian_filter(bin_statistic['statistic'], 5)
pcm = pitch.heatmap(bin_statistic, ax=ax,cmap = 'hot',edgecolors='#22312b', lw = 0)

# bin_statistic = pitch.bin_statistic(df_pressure.x, df_pressure.y, statistic='count', bins=(25, 25))
# bin_statistic['statistic'] = gaussian_filter(bin_statistic['statistic'], 1)
# pcm = pitch.heatmap(bin_statistic, ax=ax, cmap='hot', edgecolors='#22312b')
# # Add the colorbar and format off-white
# cbar = fig.colorbar(pcm, ax=ax, shrink=0.6)
# cbar.outline.set_edgecolor('#efefef')
# cbar.ax.yaxis.set_tick_params(color='#efefef')
# ticks = plt.setp(plt.getp(cbar.ax.axes, 'yticklabels'), color='#efefef')

# Add the colorbar and format off-white
cbar = fig.colorbar(pcm, ax=ax, shrink=0.6, ticks=[3, 7], pad = 0.01)
cbar.outline.set_edgecolor('#efefef')
cbar.ax.yaxis.set_tick_params(color='#efefef')
cbar.ax.set_yticklabels(['Low', 'High'], color='#efefef')
ticks = plt.setp(plt.getp(cbar.ax.axes, 'yticklabels'), color='#efefef')


# show that mcwfcs goal is the top one
pitch.text(110,1, "MCI", ax = ax, color = "w", fontsize = 15)

# add downward arrow 
pitch.arrows(108, 9, 95, 9, ax = ax, width = 2, color = "w")


# add opp in bottom right corner 
pitch.text(5, 62.5, "OPP", ax = ax, color = "w", fontsize = 15)

# add upward arrow

pitch.arrows(12, 71, 25, 71, ax = ax, width = 2, color = "w")

plt.xlabel("Created with data from StatsBomb\ngithub.com/statsbomb/open-data\n@astro_katherine", fontsize = 6, color = "w")

plt.title("Pressure applied by MCWFC\nin the 2020/21 WSL Season", color = "w")

plt.savefig("/Users/katherineormerod/Documents/football stats/mcwfc_pressure.png", dpi = 300, bbox_inches = "tight")
plt.show()
plt.close()
```
